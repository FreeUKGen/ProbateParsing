```python
from PIL import Image
import sys, os
```

```python
folder_path = "1873 Probate/1873 Probate A/"
dest_path = "1873 Probate/A2/"
folders = os.listdir(folder_path)
for files in folders:
    print(files)
    file = files.split('/')[-1]
    print(file)
    original = Image.open(folder_path+file)
    width, height = original.size
    crop_w = width - 2458
    crop_h = height - 3906
    cropped = original.crop((int(crop_w / 2)+100, int(crop_h / 2), int(width - (crop_w / 2)), int(height - (crop_h / 2))))
    cropped.save(dest_path + file)
```
